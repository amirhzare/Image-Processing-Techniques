# Histogram Processing

A histogram represents the frequency of occurrence of each pixel intensity value in an image. For grayscale images, it shows how many times each gray level appears, from black (0) to white (255). This information is crucial for tasks like image enhancement, thresholding, and segmentation.

```python
# Import necessary libraries
import cv2  
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load the image in grayscale mode
image = cv2.imread('images/01/1_1.png', 0)  

# Display the image
plt.figure(figsize=(3,3))
plt.imshow(image, cmap='gray')
plt.title('Grayscale Image')
```

```python
# Use OpenCV's calcHist function
hist_cv = cv2.calcHist([image], channels=[0], mask=None, histSize=[256], ranges=[0, 256])

# Plot the histogram
plt.figure(figsize=(10,3))
plt.title('Histogram Using OpenCV')
plt.xlabel('Pixel Intensity Value')
plt.ylabel('Number of Pixels')
plt.plot(hist_cv)
plt.xlim([0, 255])
```

```python
# Calculate histograms using NumPy
image_1_1 = cv2.imread('images/01/1_1.png')
image_1_2 = cv2.imread('images/01/1_2.png')

hist1 = cv2.calcHist([image_1_1], [0], None, [256], [0, 256])
hist2 = cv2.calcHist([image_1_2], [0], None, [256], [0, 256])

# Plot the histograms
plt.figure(figsize=(12,6))

plt.subplot(2,2,1)
plt.title('Image 1_1')
plt.imshow(image_1_1)
plt.axis('off')

plt.subplot(2,2,3)
plt.title('Histogram of Image 1_1')
plt.xlabel('Pixel Intensity Value')
plt.ylabel('Number of Pixels')
plt.plot(hist1)
plt.xlim([0, 255])

plt.subplot(2,2,2)
plt.title('Image 1_2')
plt.imshow(image_1_2)
plt.axis('off')

plt.subplot(2,2,4)
plt.title('Histogram of Image 1_2')
plt.xlabel('Pixel Intensity Value')
plt.ylabel('Number of Pixels')
plt.plot(hist2)
plt.xlim([0, 255])
```
